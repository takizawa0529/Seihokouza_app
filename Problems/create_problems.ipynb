```python
import fitz  # PyMuPDF
import csv
import re

# PDFファイルのパス
pdf_path = ".\生命保険計理\生命保険計理_2022A.pdf"
# 出力するCSVファイルのパス
csv_path = "exam_questions.csv"

# PDFを開く
doc = fitz.open(pdf_path)

# 抽出したデータを格納するリスト
data = []

# ページを順番に処理
for page_num in range(len(doc)):
    page = doc.load_page(page_num)
    text = page.get_text()

    # 問題番号、問題文、語群を抽出
    lines = text.split('\n')
    question_number = None
    question_text = []
    word_choices = []
    in_choices = False
    
    for line in lines:
        if "語群" in line:
            in_choices = True
            continue  # 語群のラベルは無視
        if re.match(r'^\[\s*\d+\s*\]', line):
            if question_number is not None:
                # 既に問題文を保存している場合、それをリストに追加
                data.append([question_number, ' '.join(question_text), '、 '.join(word_choices)])
                question_text = []
                word_choices = []
            # 問題番号を取得
            question_number = re.findall(r'\d+', line)[0]
            question_text.append(line.strip())
            in_choices = False
        elif in_choices:
            # 語群に含まれるものを追加
            word_choices.append(line.strip())
        else:
            # 問題文の続きとして扱う
            question_text.append(line.strip())

    # 最後の問題をリストに追加
    if question_number is not None:
        data.append([question_number, ' '.join(question_text), '、 '.join(word_choices)])

# CSVファイルに書き込み
with open(csv_path, 'w', newline='', encoding='utf-8 sig') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["問題番号", "問題文", "語群"])
    writer.writerows(data)

print("CSVファイルが作成されました:", csv_path)
```

```python
import fitz  # PyMuPDF
import os

subject = '生命保険計理'
exam = '2023'
alphabet = 'A'

# PDFファイルのパス
pdf_path = f".\{subject}\{subject}_{exam}{alphabet}.pdf"
# 保存する画像ファイルのパス
output_folder = f"./{subject}/{exam}_{alphabet}"
# ディレクトリの作成（必要なら）
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# PDFを開く
doc = fitz.open(pdf_path)

# 全てのページを画像として保存
for page_number in range(len(doc)):
    # ページを読み込む
    page = doc.load_page(page_number)
    
    # ページを画像として保存
    pix = page.get_pixmap()
    image_path = f"{output_folder}/page_{page_number + 1}.png"
    pix.save(image_path)
    
    print(f"ページ {page_number + 1} を画像として保存しました: {image_path}")

print("全ページの画像保存が完了しました。")
```
