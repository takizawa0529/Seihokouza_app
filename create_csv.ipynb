```python
import os
import csv

# データのテンプレート
data = [
    ['問', '問題', '語群', '解答', '解説']
]

# 50個の問題のテンプレートデータを追加
for i in range(50):
    datum = [f'{i+1}', '', '', '', '']  # '語群'と'解説'を追加
    data.append(datum)

# 年度と科目のリスト
exams = ['2023', '2022', '2021']
alphabets = ['A', 'B', 'C']
subjects_list = ['生命保険総論', '生命保険計理']

# CSVファイルの書き出し
for subject in subjects_list:
    # ディレクトリが存在しない場合は作成
    if not os.path.exists(f'./problems/{subject}'):
        os.makedirs(f'./problems/{subject}')
    
    print(subject)
    for exam in exams:
        for alphabet in alphabets:
            # CSVファイルの作成
            with open(f'problems/{subject}/{exam}_{alphabet}.csv', 'w', newline='', encoding='shift-jis') as f:
                writer = csv.writer(f)
                writer.writerows(data)
            print(f'{exam}_{alphabet}.csv : Complete')
    print('-'*40)
```
